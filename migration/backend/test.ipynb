```python
from datetime import datetime, timezone

# Get the current time in UTC, as a timezone-aware object
current_time_utc = datetime.now(timezone.utc)

# Format it to ISO 8601 string with seconds precision
iso_string = current_time_utc.isoformat(timespec='seconds')
# Result looks like '2025-11-17T14:40:00+00:00' (time will vary)

# Replace the "+00:00" with the "Z" suffix
formatted_time = iso_string.replace("+00:00", "Z")

print(f"Current time in YYYY-MM-DDTHH:MM:SSZ format: {formatted_time}")
```

```python
import jwt, json
from cryptography.hazmat.primitives import serialization
from cryptography.hazmat.primitives.asymmetric import rsa

# Load your private RSA key
with open("private_key.pem", "rb") as f:
    private_key = serialization.load_pem_private_key(f.read(), password=None)

# Extract public numbers
public_key = private_key.public_key()
numbers = public_key.public_numbers()
n = numbers.n
e = numbers.e

# Convert n and e to base64url
import base64
def b64url_uint(val):
    raw = val.to_bytes((val.bit_length() + 7) // 8, byteorder="big")
    return base64.urlsafe_b64encode(raw).rstrip(b"=").decode("utf-8")

jwk = {
    "kty": "RSA",
    "n": b64url_uint(n),
    "e": b64url_uint(e)
}

header = {
    "alg": "RS256",
    "typ": "JWT",
    "jwk": jwk
}

payload = {"sub": "example", "msg": "hello world"}

jwt_token = jwt.encode(payload, private_key, algorithm="RS256", headers=header)
print(jwt_token)
```
